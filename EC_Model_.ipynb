```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE

# Load the combined dataset (already split into training and testing sets)
X_train = pd.read_csv('/content/drive/MyDrive/EC/X_train sequence.csv')  # Already split training features
y_train = pd.read_csv('/content/drive/MyDrive/EC/y_train.csv')  # Already split training labels
X_test = pd.read_csv('/content/drive/MyDrive/EC/X_test no sequence.csv')  # Already split testing features
y_test = pd.read_csv('/content/drive/MyDrive/EC/y_test.csv')  # Already split testing labels

# ***Drop any non-numeric columns (like 'ID' or 'Sequence' if present)***
X_train = X_train.select_dtypes(include=['number'])  # Keep only numeric columns
X_test = X_test.select_dtypes(include=['number'])  # Keep only numeric columns for test data

# Apply SMOTE to oversample the minority class in the training set
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train['Activity']) # Pass y_train['Activity'] instead of flattened y_train

# Print the resampled data class distribution
print(f"Resampled data class distribution:\n{pd.Series(y_res).value_counts()}") # Use pd.Series to apply value_counts()

# Initialize the RandomForestClassifier
model = RandomForestClassifier(random_state=42, class_weight='balanced')

# Train the model using the resampled data
model.fit(X_res, y_res)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test['Activity'], y_pred) # Access 'Activity' column for accuracy calculation
classification_rep = classification_report(y_test['Activity'], y_pred) # Access 'Activity' column for classification report

print(f"Accuracy on Test Data (Random Forest): {accuracy}")
print(f"Classification Report on Test Data (Random Forest):\n", classification_rep)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, roc_auc_score, average_precision_score, matthews_corrcoef, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier

# Load the resampled training data
X_train = pd.read_csv('/content/drive/MyDrive/EC/X_train_resampled.csv')  # Resampled features
y_train = pd.read_csv('/content/drive/MyDrive/EC/y_train_resampled.csv')  # Resampled labels

# Load the test data
X_test = pd.read_csv('/content/drive/MyDrive/EC/X_test no sequence.csv')  # Test features
y_test = pd.read_csv('/content/drive/MyDrive/EC/y_test.csv')  # Test labels

# Define models for comparison
models = {
    "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced'),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "SVM (RBF Kernel)": SVC(kernel='rbf', random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "Perceptron": Perceptron(random_state=42),
    "Neural Network (MLP)": MLPClassifier(random_state=42)
}

# Initialize cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define a function to calculate specificity
def specificity(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return tn / (tn + fp)

# Initialize dictionaries to store metrics
metrics = {
    "Model": [],
    "Precision": [],
    "Recall": [],
    "Specificity": [],
    "AUC-PR": [],
    "MCC": []
}

# Train and evaluate each model using cross-validation
for model_name, model in models.items():
    print(f"\nTraining and evaluating {model_name}...")

    # Lists to store metric results during cross-validation
    precision_list = []
    recall_list = []
    specificity_list = []
    auc_pr_list = []
    mcc_list = []

    # Perform cross-validation
    for train_idx, val_idx in cv.split(X_train, y_train):
        X_train_cv, X_val_cv = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_train_cv, y_val_cv = y_train.iloc[train_idx], y_train.iloc[val_idx]

        # Train the model
        model.fit(X_train_cv, y_train_cv.values.ravel())

        # Make predictions
        y_pred = model.predict(X_val_cv)

        # Calculate metrics for the fold
        precision_list.append(precision_score(y_val_cv, y_pred))
        recall_list.append(recall_score(y_val_cv, y_pred))
        specificity_list.append(specificity(y_val_cv, y_pred))
        auc_pr_list.append(average_precision_score(y_val_cv, y_pred))
        mcc_list.append(matthews_corrcoef(y_val_cv, y_pred))

    # Store the mean metrics for the model
    metrics["Model"].append(model_name)
    metrics["Precision"].append(np.mean(precision_list))
    metrics["Recall"].append(np.mean(recall_list))
    metrics["Specificity"].append(np.mean(specificity_list))
    metrics["AUC-PR"].append(np.mean(auc_pr_list))
    metrics["MCC"].append(np.mean(mcc_list))

# Convert metrics to a DataFrame for easier plotting
metrics_df = pd.DataFrame(metrics)

# Plot the comparison of models based on different metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Precision comparison
metrics_df.plot(kind='bar', x='Model', y='Precision', ax=axes[0, 0], color='b', legend=False)
axes[0, 0].set_title('Precision Comparison')
axes[0, 0].set_ylabel('Precision')

# Recall comparison
metrics_df.plot(kind='bar', x='Model', y='Recall', ax=axes[0, 1], color='r', legend=False)
axes[0, 1].set_title('Recall Comparison')
axes[0, 1].set_ylabel('Recall')

# Specificity comparison
metrics_df.plot(kind='bar', x='Model', y='Specificity', ax=axes[1, 0], color='g', legend=False)
axes[1, 0].set_title('Specificity Comparison')
axes[1, 0].set_ylabel('Specificity')

# AUC-PR comparison
metrics_df.plot(kind='bar', x='Model', y='AUC-PR', ax=axes[1, 1], color='purple', legend=False)
axes[1, 1].set_title('AUC-PR Comparison')
axes[1, 1].set_ylabel('AUC-PR')

plt.tight_layout()
plt.show()

# Print the metrics for each model
print("\nComparison of models with key metrics:")
print(metrics_df)
```

```python
!pip install biopython
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from Bio import pairwise2 # This line should work after installation
from tqdm import tqdm

# Load your data
combined_data = pd.read_csv('/content/drive/MyDrive/EC_data/combined_data  EC(27).csv')

# Shuffle the dataset first
combined_data = shuffle(combined_data, random_state=42).reset_index(drop=True)

# Print first few rows to verify
print("Combined Data (First 5 rows):\n", combined_data.head())

# Check the class distribution (assuming 'Activity' column indicates active (1) and inactive (0))
print("\nClass distribution:\n", combined_data['Activity'].value_counts())

# Separate features and target variable
X = combined_data.drop(columns=['Activity', 'ID'])  # Features (Amino acid columns)
y = combined_data['Activity']  # Target variable (Activity)

# Define function to calculate sequence identity
def sequence_identity(seq1, seq2):
    alignment = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)[0]
    matches = alignment[2]  # number of matches
    length = max(len(seq1), len(seq2))
    return matches / length

# Initialize lists for train and test indices
train_indices = []
test_indices = []

# **Get the number of unique IDs from the 'ID' column.**
num_ids = len(combined_data['ID'].unique())

# Set target test size using the number of unique IDs
target_test_size = int(0.1 * num_ids)  # 10% of total data

print("Splitting with ≤40% similarity and target 10% test size...")

# **Get sequences for similarity comparison**
sequences = combined_data['Sequence']  # Use 'Sequence' column for similarity check

# Initialize test count
test_count = 0

# Iterate through sequences and assign to train or test sets
for i in tqdm(range(len(sequences)), desc="Processing sequences"):
    is_similar = False
    # Check if current sequence is similar to any in the test set
    for j in test_indices:
        if sequence_identity(sequences[i], sequences[j]) > 0.4:  # Check for over 40% similarity
            is_similar = True
            break

    # Add to test set if not similar to others and within size limit
    if not is_similar and test_count < target_test_size:
        test_indices.append(i)
        test_count += 1
    else:
        train_indices.append(i)

# Final split
X_train = X.loc[train_indices]
X_test = X.loc[test_indices]
y_train = y.loc[train_indices]
y_test = y.loc[test_indices]

# Report the sizes of train and test sets
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
print(f"Test size percentage: {X_test.shape[0] / len(sequences) * 100:.2f}%")

# Save if needed
X_train.to_csv('/content/X_train.csv', index=False)
X_test.to_csv('/content/X_test.csv', index=False)
y_train.to_csv('/content/y_train.csv', index=False)
y_test.to_csv('/content/y_test.csv', index=False)

print("✅ Train/test split completed!")
```
