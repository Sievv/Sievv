```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.impute import SimpleImputer

X_train = pd.read_csv('/content/drive/MyDrive/PS/X_train v2.csv')
y_train = pd.read_csv('/content/drive/MyDrive/PS/y_train v1.csv').squeeze()

X_test = pd.read_csv('/content/drive/MyDrive/PS/X_test v3.csv')
y_test = pd.read_csv('/content/drive/MyDrive/PS/y_test.csv').squeeze()

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Drop non-feature columns like 'ID' and 'Sequence' if present
drop_cols = ['ID', 'Sequence']
X_train = X_train.drop(columns=[col for col in drop_cols if col in X_train.columns])
X_test = X_test.drop(columns=[col for col in drop_cols if col in X_test.columns])

# Align columns between train and test
X_test = X_test[X_train.columns]  # Ensure same column order

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)

# Define models
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store results
results = []

# ROC Curve setup
plt.figure(figsize=(10, 8))

for name, model in models.items():
    print(f"Evaluating {name}...")

    # Cross-validated predictions
    y_pred = cross_val_predict(model, X_train, y_train, cv=cv)
    y_proba = cross_val_predict(model, X_train, y_train, cv=cv, method='predict_proba')[:, 1]

    # Evaluation metrics
    acc = accuracy_score(y_train, y_pred)
    cm = confusion_matrix(y_train, y_pred)
    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    auc_score = roc_auc_score(y_train, y_proba)

    results.append({
        'Model': name,
        'Accuracy': acc,
        'Sensitivity': sensitivity,
        'Specificity': specificity,
        'AUC': auc_score
    })

    # Plot ROC
    fpr, tpr, _ = roc_curve(y_train, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc_score:.2f})")

# Finalize ROC
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison (5-fold CV)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Show results
results_df = pd.DataFrame(results).sort_values(by='AUC', ascending=False).reset_index(drop=True)
print("\nModel Comparison (CV Results):")
print(results_df)
```

```python
print("\nTest Set Evaluation with Best Model:")
best_model_name = results_df.iloc[0]['Model']
best_model = models[best_model_name]

best_model.fit(X_train, y_train)
test_preds = best_model.predict(X_test)
test_proba = best_model.predict_proba(X_test)[:, 1]

# Test set metrics
acc = accuracy_score(y_test, test_preds)
cm = confusion_matrix(y_test, test_preds)
tn, fp, fn, tp = cm.ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
auc_score = roc_auc_score(y_test, test_proba)

print(f"Accuracy: {acc:.2f}, Sensitivity: {sensitivity:.2f}, Specificity: {specificity:.2f}, AUC: {auc_score:.2f}")
```

```python
from sklearn.inspection import permutation_importance

# Function to plot feature importance
def plot_feature_importance(importances, model_name):
    importance_df = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': importances
    }).sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
    plt.xlabel('Importance')
    plt.title(f'Feature Importance - {model_name}')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()


# Loop through models
for name, model in models.items():
    print(f"Processing {name}...")

    # Apply imputation before fitting the model
    imputer = SimpleImputer(strategy='mean')
    X_train_imputed = imputer.fit_transform(X_train) # Impute missing values
    X_train_imputed = pd.DataFrame(X_train_imputed, columns=X_train.columns) # Keep as DataFrame

    model.fit(X_train_imputed, y_train) # Fit with imputed data

    if name == 'SVM':
        # Use permutation importance for SVM
        result = permutation_importance(model, X_train_imputed, y_train, n_repeats=10, random_state=42) # Use imputed data
        importances = result.importances_mean
    else:
        # Use built-in feature importance
        importances = model.feature_importances_

    # Plot
    plot_feature_importance(importances, name)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
    precision_recall_curve,
    auc,
    matthews_corrcoef,
)

# === Step 1: Load training and test data ===
X_train = pd.read_csv('/content/drive/MyDrive/PS/X_train v2.csv')
y_train = pd.read_csv('/content/drive/MyDrive/PS/y_train v1.csv').squeeze()
X_test = pd.read_csv('/content/drive/MyDrive/PS/X_test v3.csv')
y_test = pd.read_csv('/content/drive/MyDrive/PS/y_test.csv').squeeze()

# === Step 2: Preprocessing ===
drop_cols = ['ID', 'Sequence']
X_train = X_train.drop(columns=[col for col in drop_cols if col in X_train.columns])
X_test = X_test.drop(columns=[col for col in drop_cols if col in X_test.columns])
X_test = X_test[X_train.columns]  # Ensure same order

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)

# === Step 3: Define and Evaluate Models with Cross-Validation ===
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    print(f"\n=== {name} Cross-Validation Results ===")

    # Cross-validation setup
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Get predicted probabilities and labels
    y_pred_cv = cross_val_predict(model, X_train, y_train, cv=cv, method='predict')
    y_prob_cv = cross_val_predict(model, X_train, y_train, cv=cv, method='predict_proba')[:, 1]

    # Metrics
    acc = accuracy_score(y_train, y_pred_cv)
    auc_roc = roc_auc_score(y_train, y_prob_cv)
    precision, recall, _ = precision_recall_curve(y_train, y_prob_cv)
    auc_pr = auc(recall, precision)
    mcc = matthews_corrcoef(y_train, y_pred_cv)

    tn, fp, fn, tp = confusion_matrix(y_train, y_pred_cv).ravel()
    specificity = tn / (tn + fp)
    sensitivity = tp / (tp + fn)

    results[name] = {
        'Accuracy': acc,
        'ROC AUC': auc_roc,
        'PR AUC': auc_pr,
        'MCC': mcc,
        'Specificity': specificity,
        'Sensitivity': sensitivity
    }

    print(f"Accuracy       : {acc:.3f}")
    print(f"ROC AUC        : {auc_roc:.3f}")
    print(f"PR AUC         : {auc_pr:.3f}")
    print(f"MCC            : {mcc:.3f}")
    print(f"Specificity    : {specificity:.3f}")
    print(f"Sensitivity    : {sensitivity:.3f}")

    # Plot ROC and PR curves
    fpr, tpr, _ = roc_curve(y_train, y_prob_cv)
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f'AUC = {auc_roc:.2f}')
    plt.plot([0, 1], [0, 1], '--', color='gray')
    plt.title(f'{name} - ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(recall, precision, label=f'PR AUC = {auc_pr:.2f}')
    plt.title(f'{name} - Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()

    plt.tight_layout()
    plt.show()

results

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# === Step 1: Load training and test data ===
X_train = pd.read_csv('/content/drive/MyDrive/PS/X_train v2.csv')
y_train = pd.read_csv('/content/drive/MyDrive/PS/y_train v1.csv').squeeze()
X_test = pd.read_csv('/content/drive/MyDrive/PS/X_test v3.csv')
y_test = pd.read_csv('/content/drive/MyDrive/PS/y_test.csv').squeeze()

# === Step 2: Preprocessing ===
drop_cols = ['ID', 'Sequence']
X_train = X_train.drop(columns=[col for col in drop_cols if col in X_train.columns])
X_test = X_test.drop(columns=[col for col in drop_cols if col in X_test.columns])
X_test = X_test[X_train.columns]  # Ensure same order

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)

# === Step 3: Define and Train Classifiers ===
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"\n=== {name} Validation Results ===")
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("ROC AUC:", roc_auc_score(y_test, y_prob))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    # Optional: Pie chart of test predictions
    counts = pd.Series(y_pred).value_counts().sort_index()
    labels = ['Inactive (0)', 'Active (1)']
    sizes = [counts.get(0, 0), counts.get(1, 0)]
    colors = ['#ff9999','#66b3ff']

    plt.figure(figsize=(5, 5))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
    plt.axis('equal')
    plt.title(f'{name} Prediction Distribution (Test Set)')
    plt.tight_layout()
    plt.show()
```
